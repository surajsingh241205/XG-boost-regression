# Explanation of XGBoost

**Simple definition:**
- XGBoost is an advanced version of Gradient Boosting.
- It builds many small decision trees one by one, and each tree tries to fix the mistakes of the previous tree.

Why XGBoost is popular:
- Very accurate
- Fast
- Works well on real datasets
- Used in Kaggle & real companies

Simple line:
- Gradient Boosting = strong
- XGBoost = stronger + faster + optimized

## Real Life Analogy

**Think like this:**
You are learning math.

- First try → many mistakes
- Second try → correct mistakes
- Third try → improve more
- Fourth try → almost perfect

Each try improves the previous one.

That is XGBoost.
- Final result = combination of all corrections

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

from xgboost import XGBRegressor


# dataset
np.random.seed(42)

X = np.random.rand(1000, 1) * 10
y = 5 * X.squeeze() + 3 + np.random.randn(1000) * 2

df = pd.DataFrame({
    "X": X.squeeze(),
    "y": y
})


# split
X_train, X_test, y_train, y_test = train_test_split(
    df[["X"]],
    df["y"],
    test_size=0.2,
    random_state=42
)


# model
model = XGBRegressor()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)


print("R2 Score:", r2_score(y_test, y_pred))
```

```python
df.to_csv('XG boost synthetic dataset')
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, label="Actual")
plt.scatter(X_test, y_pred, color="red", label="Predicted")

plt.legend()
```
